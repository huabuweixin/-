```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

df=pd.read_csv('Slopecharts_Data1.csv')

left_label = [str(c) + ',' + str(round(y)) for c, y in zip(df.Contry, df['1970'])]
right_label = [str(c) + ',' + str(round(y)) for c, y in zip(df.Contry, df['1979'])]

def newline(p1,p2,color='black'):
    ax=plt.gca()
    l = mlines.Line2D([p1[0],p2[0]],[p1[1],p2[1]],color='red' if p1[1]-p2[1] >0 else 'blue',marker='o',markersize=5)
    ax.add_line(l)
    return l

#创建空白图
fig, ax = plt.subplots(1, 1, figsize=(6,12), dpi=80)
ax.set_xticks([])  # 隐藏x轴刻度
ax.set_yticks([])  # 隐藏y轴刻度
#绘制两根垂线
ax.vlines(x=1,ymin=40,ymax=90,color='black',alpha=1,linewidth=1.2,linestyles='solid')
ax.vlines(x=3,ymin=40,ymax=90,color='black',alpha=1,linewidth=1.2,linestyles='solid')
ax.text(1, 90, '1970', ha='center', va='bottom')
ax.text(3, 90, '1979', ha='center', va='bottom')
#绘制点
ax.scatter(y=df['1970'],x=np.repeat(1,df.shape[0]),s=10,color='grey',alpha=1)
ax.scatter(y=df['1979'],x=np.repeat(3,df.shape[0]),s=10,color='grey',alpha=1)
for p1,p2,c,in zip(df['1970'],df['1979'],df['Contry']):
    newline([1,p1],[3,p2])
    ax.text(1-0.05,p1,c+','+str(round(p1))
    ,horizontalalignment='right'
    ,verticalalignment = 'center'
    ,fontdict={'size':9})
    ax.text(3+0.05,p2,c+', '+str(round(p2))
    ,horizontalalignment='left'
    ,verticalalignment = 'center' 
    ,fontdict={'size':9})


plt.gca().spines['top'].set_alpha(.0)
plt.gca().spines['bottom'].set_alpha(.0)
plt.gca().spines['right'].set_alpha(.0)
plt.gca().spines['left'].set_alpha(.0)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

df = pd.read_csv('Slopecharts_Data2.csv')

years = [2007, 2008, 2009, 2010, 2011, 2012, 2013]
df['color'] = df.apply(lambda row: 'blue' if row['2013'] > row['2007'] else 'red', axis=1)

fig, ax = plt.subplots(1, 1, figsize=(6, 12), dpi=80)
ax.set_xticks([])  # 隐藏x轴刻度
ax.set_yticks([])  # 隐藏y轴刻度

ax.vlines(x=2007, ymin=-100000, ymax=2800000, color='black', alpha=1, linewidth=1.2, linestyles='solid')
ax.vlines(x=2013, ymin=-100000, ymax=2800000, color='black', alpha=1, linewidth=1.2, linestyles='solid')
ax.text(2007, 2800000, '2007', ha='center', va='bottom')
ax.text(2013, 2800000, '2013', ha='center', va='bottom')

ax.scatter(y=df['2007'], x=np.repeat(2007, df.shape[0]), s=10, color='grey', alpha=1)
ax.scatter(y=df['2013'], x=np.repeat(2013, df.shape[0]), s=10, color='grey', alpha=1)
for i, row in df.iterrows():
    # 绘制折线
    ax.plot(years, [row[str(year)] for year in years], color=row['color'], linestyle='-',label=row['continent'])
    ax.text(2007-0.05,row['2007'],f"{row['continent']}, {row['2007']}"
    ,horizontalalignment='right'
    ,verticalalignment = 'center'
    ,fontdict={'size':9})
    ax.text(2013+0.05,row['2013'],f"{row['continent']}, {row['2013']}"
    ,horizontalalignment='left'
    ,verticalalignment = 'center' 
    ,fontdict={'size':9})
    for year in [2007, 2013]:
        ax.plot(year, row[str(year)], 'o', markersize=5, markeredgewidth=1, markeredgecolor="black",color="white")



plt.gca().spines['top'].set_alpha(.0)
plt.gca().spines['bottom'].set_alpha(.0)
plt.gca().spines['right'].set_alpha(.0)
plt.gca().spines['left'].set_alpha(.0)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('Data3.csv')


angles=np.arange(0,2*np.pi,2*np.pi/df.shape[0])
radius=np.array(df['Price'])

fig=plt.figure(figsize=(4,4))
fig=plt.subplot(projection='polar')
fig.set_theta_offset(np.pi/4)
fig.set_theta_direction(-1)
fig.set_rlabel_position(-45)    
plt.bar(angles,radius,color="#f7a873")
plt.xticks(angles,df['Day'])

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
df=pd.read_csv('Data4.csv')
df['Total'] = df['Peter'] + df['Jack'] + df['Eelin']

dfsorted = df.sort_values(by='Total', ascending=True)

days = dfsorted['Day']
peter_values = dfsorted['Peter']
jack_values = dfsorted['Jack']
eelin_values = dfsorted['Eelin']

angles=np.arange(0,2*np.pi,2*np.pi/df.shape[0])

fig=plt.figure(figsize=(8,8))
fig=plt.subplot(projection='polar')
fig.set_theta_offset(np.pi/4)
fig.set_theta_direction(1)
fig.set_rlabel_position(-32)  
fig.set_yticks([50, 100, 150, 200])
fig.set_ylim(0, 220) 
fig.set_theta_offset(5*np.pi / 7 )
fig.grid(True)
cmap = cm.get_cmap('Blues')  
color1 = cmap(0.2) 
color2 = cmap(0.4)  
color3 = cmap(0.6) 

plt.bar(angles, peter_values, width=2 * np.pi / len(days), edgecolor='black', color=color1, label='Peter')
plt.bar(angles, jack_values, width=2 * np.pi / len(days), bottom=peter_values, edgecolor='black', color=color2, label='Jack')
plt.bar(angles, eelin_values, width=2 * np.pi / len(days), bottom=peter_values + jack_values, edgecolor='black', color=color3, label='Eelin')
plt.xticks(angles,df['Day'])

plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
