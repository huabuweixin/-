```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
df=pd.read_csv('data.csv')

result = df.iloc[0:, 0].tolist()
N=len(result)

# 设置每个点的角度值
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# 初始化极坐标网格
ax = plt.subplot(111, polar=True)

# 设置角度偏移
ax.set_theta_offset(pi / 2)
# 设置顺时针还是逆时针，1或者-1
ax.set_theta_direction(-1)

# 设置x轴的标签
plt.xticks(angles[:-1], result)

# Draw ylabels
# 画标签
ax.set_rlabel_position(0)
plt.yticks([10, 20, 30], ["10", "20", "30"], color="grey", size=7)
plt.ylim(0, 40)

values =  df.iloc[0:, 1].tolist()
values += values[:1]
ax.plot(angles, values,marker='o', linewidth=1, linestyle='solid', label="group A",color="#3118eb")
ax.fill(angles, values, 'b', alpha=0.1)

# Ind2
values =  df.iloc[0:, 2].tolist()
values += values[:1]
ax.plot(angles, values, marker='o',linewidth=1, linestyle='solid', label="group B",color="#eb5119")
ax.fill(angles, values, 'r', alpha=0.1)


# Add legend
# 添加图例
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```

```python
import matplotlib.pyplot as plt

from wordcloud import WordCloud
with open("WordCloud.txt",'r') as f:
    s = f.read()
wordcloud = WordCloud(
    width=800,
    height=800,
    background_color='white',
    colormap='Oranges',
    max_font_size=90,
    min_font_size=30

).generate(s)
 
# 展示词云图
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.show()
```
